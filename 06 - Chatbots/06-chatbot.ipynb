Questão 55 - Crie um chatbot baseado em regras para uma pizzaria, o chatbot deve fazer a saudação, coletar o pedido e as informações de entrega. Em seguida o chabot confirma o pedido e se despede do usuário.

```python
def fazer_pedido():
    Pizzas_disponiveis = ["Calabresa", "Portuguesa", "Margherita", "Frango com catupiry", "Muçarela"]
    print("Pizzas disponíveis:")
    valor_pizza=[23, 25, 24, 26, 20]
 

    for pizza, valor in zip(Pizzas_disponiveis, valor_pizza):
        print(f"{Pizzas_disponiveis.index(pizza)+1}. {pizza} {valor} reais")

    escolha = int(input("Digite qual sabor de pizza vai querer: "))
    
    quantidade = int(input("Digite quantas pizza vai querer:"))
    
    total = valor_pizza[escolha-1] * quantidade
    try:
        escolha = int(escolha)
        if 1 <= escolha <= len(Pizzas_disponiveis):
            horario_escolhido = Pizzas_disponiveis[escolha - 1]
            print(f"Você selecionou a Pizza: {horario_escolhido}.")
        else:
            print("Escolha inválida. Por favor, digite um número válido.")
    except ValueError:
        print("Escolha inválida. Por favor, digite um número válido.")
    
    return int(total)
```

```python
def checkout(total):
    print(f"seu pedido deu um total de {total} Reais.")
    destino = input("Digite qual o seu endereço: ")
    print("Obrigado por usar nosso serviço. Adeus!")
```

```python
def bob_pizza():
    print("Bem-vindo a Pizzaria do Bob!")

    valor_total = 0

    while True:
        print("\nOpções:")
        print("1. Fazer um pedido")
        print("2. Checkout")
        print("3. Sair")


        comando = input("Digite o número da opção desejada: ")



        if comando == "1":
            valor_total=fazer_pedido()

        elif comando == "2":
            checkout(valor_total)
            break
        elif comando == "3":
            print("Obrigado por usar nosso serviço. Adeus!")
            break
        else:
            print("Opção inválida. Por favor, digite um número válido.")
```

```python
def stop_words(text):
    doc = nlp(text)
    
    no_stop_words = [token.text for token in doc if not token.is_stop and not token.is_punct]
    return " ".join(no_stop_words)
```

```python
from nltk.stem import RSLPStemmer
def stemming(text):
    stemmer = RSLPStemmer()
    stem_words = [stemmer.stem(word) for word in text]
    return "".join(stem_words)
```

```python
def pipeline(text):
    text_without_stop_words=stop_words(text)
    text_after_stemming=stemming(text_without_stop_words)
    return text_after_stemming
```

```python
df["nova_Frase"] = df["Frase"].apply(pipeline)
```
