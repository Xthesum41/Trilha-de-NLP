```python
def sentiment_analyzer(text):
    positive_count=0
    regex_text = re.findall("\w+",text)
    for linha in positive_words:
        linha = linha.strip('\n')
        for positive in regex_text:
            if linha == positive:
                positive_count += 1
    negative_count=0
    for linha in negative_words:
        linha = linha.strip('\n')
        for negative in regex_text:
            if linha == negative:
                negative_count += 1
    if (positive_count > negative_count):
        print("positive text")
    else:
        print("negative text")
```

```python
sentiment_analyzer(text1)
```

```python
sentiment_analyzer(text2)
```
