```python
from nltk.stem import RSLPStemmer
```

```python
def stemming(text):
    stemmer = RSLPStemmer()
    stem_words = [stemmer.stem(word) for word in text]
    return stem_words
#tokenize_words = word_tokenize(text)
stemming(word_tokenize('Os gatos estavam pulando no telhado e os cães latindo no quintal'))
```

Questão 11

O processo de stemming busca remover os sufixos de uma palavra para obter sua raiz e resultado do stemming nem sempre é uma palavra real, e muitas vezes é uma forma truncada da palavra.
Ja a lematização, por outro lado, procura obter a forma canônica ou "lema" de uma palavra, levando em consideração seu significado e contexto e o resultado da lematização é sempre uma palavra real que faz parte do vocabulário do idioma.

```python
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize,sent_tokenize
```

```python
stop_words = set(stopwords.words("english"))
```

```python
text = """Welcome you to programing knowledge. Lets start with our first tutorial on NLTK. we shall learn the basics of nltk here."""
tokenize_words = word_tokenize(text)
print(tokenize_words)
```

```python
tokenize_words_without_stop_words = []
for word in tokenize_words:
    if word not in stop_words:
        tokenize_words_without_stop_words.append(word)
print(tokenize_words_without_stop_words)
```

```python
import spacy

from spacy.lang.en.stop_words import STOP_WORDS

len(STOP_WORDS)
```

```python
nlp = spacy.load("en_core_web_sm")
```

```python
def preprocess(text):
    doc = nlp(text)
    
    no_stop_words = [token.text for token in doc if not token.is_stop and not token.is_punct]
    no_stop_words_lowered = [text.lower() for text in no_stop_words]
    new_no_stop_words_lowered = [token for token in no_stop_words_lowered if not token.isnumeric()]
    return new_no_stop_words_lowered
```

```python
text = [
    "After Narrowly 23 and 3 escaping a bizarre accident, a troubled teenager is plagued by visions of a man in a large rabbit suit who manipulates him to commit a series of crimes.",
    "On the hottest day of the year on a street in the Bedford-Stuyvesant section of Brooklyn 99, everyone's hate and bigotry smolders and builds until it explodes into violence.",
    "A New York City police officer tries to save his estranged wife and several others taken hostage by terrorists during a Christmas party at the Nakatomi Plaza in Los Angeles."
]
bla = [preprocess(token) for token in text]
bla
```
